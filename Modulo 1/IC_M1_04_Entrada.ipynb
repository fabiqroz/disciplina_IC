A função `input()` é, portanto, uma função nativa do python cujo propósito é  receber uma entrada do usuário e retorná-la para o programa.

Todo esse vocabulário sobre *funções* e *retornos* só serão destrinchados ao final do curso. Por hora vamos entender que o comando `input()` serve para **receber algo do usuário**, cujo **valor**, por sua vez será armazenado em uma **varável** como segue no exemplo abaixo, em que **a variável nome, recebe o valor lido pelo comando `input()` a partir do usuário.** 

```python
# lendo uma string digitada pelo usuario e guardando seu valor em uma variavel
nome = input('Digite o seu nome: ')
```

```python
# check
nome
```

```python
# lendo um valor inteiro a partir do usuario
num = input('Digite um número inteiro: ')
```

Perceba que o comando `input()` sempre retorna uma string. Se o objetivo for receber um valor numérico ou lógico, precisaremos converter a string para esses tipos de dados!  

```python
num = int(num)
```

```python
num
```

```python
# lendo um numero de ponto flutuante  a partir do usuario
num = float(input('Digite um número real: '))
```

```python
# check
num
```

Você também pode converter o tipo do dado utilizando o comando (função nativa do python) *eval()*, que permite que você avalie **expressões arbitrárias** do Python a partir de uma **entrada baseada em string**. 

Esta função pode ser útil quando você está tentando avaliar dinamicamente expressões Python de qualquer entrada que venha como uma string ou um objeto de código compilado.

```python
num = eval(input('Digite um número real: '))
```

```python
type(num)
```

Multiplas entradas por linha:

Entender essa sintaxe é importante, ela condensa vários conceitos vistos até agora. Vamos interpretar essa linha de código de dentro pra fora: 



1.   Primeiro comando:  `input('Digite dois valores inteiro separados por espaço: ')`

> Este comando irá retornar uma *string*, afinal, o retorno de um `input()` é sempre uma *string*. 

2.   Segundo comando: `input('Digite dois valores inteiro separados por espaço: ').split())`


> Este comando (função `split()`, nativa da classe string, já conhecida por nós) irá "pegar" essa string e subdividi-la em substrings separadas por espaço. Logo, se o usuário digitar algo como `1 4` como entrada. O `split()` ir;á quebrar a string '1 4' em duas substrings '1'e '4'.

3. Terceiro comando: `map(int, input('Digite dois valores inteiro separados por espaço: ').split())`



> este comando `(map(int, ...)`) irá **mapear** as strings '1'e '4' resultantes da quebra anterior para valores do tipo inteiro, e por fim irá retornar essas valores para as minhas variáveis `a` e `b`. 


```python
a, b = map(float, input("Digite dois numeros reais: ").split(','))
```

```python
a
```

```python
b
```

O comando (função nativa do python) para implementar a saída de dados é o `print()`. Este comando exibe uma mensagem (*string*) específica na tela. Essa mensagem pode conter valores de variáveis e pode ser formatada utilizando-se a função nativa da classe string `format()`(também já conhecia por nós). 

```python
print('Oi mundo!')
```

```python
a = 6
```

```python
print('Oi'*a)
```

```python
print('No exemplo acima, a string "Oi" foi replicada {} vezes'.format(a))
```

```python
nome = input('Qual o seu nome? ')
idade = eval(input('Qual a sua idade? '))
peso = eval(input('Digite o seu peso (em kg): '))
altura = eval(input('Digite a sua altura (em m): '))
```

```python
imc = peso/(altura**2)
```

```python
print('{0:5} | {1:5} | {2:5}'.format('Idade', 'Peso', 'Altura'))
print('{0:5} | {1:5.2f} | {2:5.2f}'.format(idade, peso, altura))
print('\n')
print('Ola, {0}. De acordo com seus dados, seu IMC é de {1:.2f}'.format(nome, imc))
```
