# Decisão

As instruções `if` em Python nos permitem executar ações alternativas com base em um determinado conjunto de resultados.

    1.) Decisão Simples (if)
    2.) Decisão Composta (if/else)
    3.) Decisão Aninhada (elif)



# Decisão Simples

"Ei, se essa condição for satisfeita, execute esta ação (ou estas ações)"


<img src = "https://drive.google.com/uc?id=1KaM19SF4ModosiHavqI2j9JU6aeJ-K8t" width = "500">


```python
condicao = True
```

```python
if condicao:
  print('A condicao eh verdadeira!!!')
```

### Exemplos com expressões lógico-relacionais:

```python
if 2>1:
  print('A condicao eh verdadeira!!!')
```

```python
a = 4
if a == 4:
  print('A condicao eh verdadeira. A variável "a" é igual a 4')
```

```python
nome = 'maria'
if nome == 'maria':
  print('Oi Maria!')
```

```python
a = 1
b = 2
c = 3

if a != c and b == 2:
  print('A condicao eh verdadeira!!!')
```

# Decisão Composta 

"Ei, se essa condição for satisfeita, execute esta ação (ou estas ações). Senão, execute aquela ação (ou aquelas ações)"


<img src = "https://drive.google.com/uc?id=1infi19-znsnik7elezrrFSEq2BcQ_8D8" width = "600">

```python
x = False
```

```python
if x:
  print('x é Verdadeiro!')
else:
  print('x não é verdadeiro!')
```

```python
nome = input()

if nome == 'Sammy':
  print('Bem vinda(o), Sammy!')
else:
  print('Bem vinda(o), {}!'.format(nome))
```

# Decisão aninhada

"Ei, se essa condição for satisfeita, execute esta ação (ou estas ações). Senão, se esta outra condição foir satisfeita, execute aquela ação (ou aquelas ações). Senão, execute aquela outra ação (ou aquelas outras ações)"

<img src = "https://drive.google.com/uc?id=1AMTt6VTfWMtpkcK671P5Q6sMm3_C0ao6" width = "300">

```python
pessoa = input()
genero = input()

if genero == 'feminino':
  print('Bem vinda {}!'.format(pessoa))
else:
  if genero == 'masculino':
    print('Bem vindo {}!'.format(pessoa))
  else:
    if genero == 'naobinario':
      print('Bem vindx {}!'.format(pessoa))
    else:
      print('Bem vinda(o)! Qual o seu nome?')
```

As instruções <code>elif</code> em Python nos permitem implementar estruturas aninhadas de forma mais organizada!

```python
nome = input()
genero = input()
genero = genero.upper()

if genero == 'FEMININO':
  print('Bem vinda, {}!'.format(nome))
elif genero == 'MASCULINO':
  print('Bem vindo, {}!'.format(nome))
elif genero == 'NAOBINARIO':
  print('Bem vindx, {}!'.format(nome))
else:
  print('Bem vindx!!')
```
